## найти все апи-схемы в нашей системе и принудительно обновить

```python
import requests
import time
import json

# —————————————————————————
# Настройки
TOKEN = "gb_api_vVVR78hJXh8dkcFDRtgn9LBHtFTQUmWAkQDdXe8q"
ORG_ID = "isNk5xPnJebTWJa4JCg5"

BASE_URL = "https://api.gitbook.com/v1"

HEADERS = {
    "Authorization": f"Bearer {TOKEN}",
    "Content-Type": "application/json",
    "Accept": "application/json",
}

# —————————————————————————
def list_openapi_specs():
    url = f"{BASE_URL}/orgs/{ORG_ID}/openapi?limit=1000"
    resp = requests.get(url, headers=HEADERS)
    data = resp.json()
    print(json.dumps(data, indent=2, ensure_ascii=False))
    resp.raise_for_status()
    return resp.json().get("items", [])

def update_spec(spec_slug, source_url):
    url = f"{BASE_URL}/orgs/{ORG_ID}/openapi/{spec_slug}"
    body = {
        "source": {
            "url": source_url
        }
    }
    resp = requests.put(url, json=body, headers=HEADERS)
    resp.raise_for_status()
    return resp.json()

def main():
    specs = list_openapi_specs()

    total = len(specs)
    updated = 0
    failed = 0
    skipped = 0

    print(f"Found {total} OpenAPI specs\n")

    for spec in specs:
        slug = spec.get("slug")
        source_url = spec.get("sourceURL")

        if not slug or not source_url:
            skipped += 1
            print(f"Skipped: missing slug or sourceURL → {spec.get('id', 'unknown')}")
            continue

        print(f"Updating: {slug}")

        try:
            update_spec(slug, source_url)
            updated += 1
            print(f" ✓ Updated")
        except Exception as e:
            failed += 1
            print(f" ✗ Failed: {e}")

        # защита от rate-limit
        time.sleep(0.2)

    # —————————————————————————
    print("\n====== SUMMARY ======")
    print(f"Total specs      : {total}")
    print(f"Updated          : {updated}")
    print(f"Failed updates   : {failed}")
    print(f"Skipped          : {skipped}")
    print("=====================")


if __name__ == "__main__":
    main()
```
