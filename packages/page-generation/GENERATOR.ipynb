```python
import requests
import json

# ==== INPUTS ====
TEMPLATE_INDEX = 7
TEMPLATES = {
    0: "[chat]-completions.md",
    1: "[chat]-responses.md",
    2: "[chat]-completions-and-responses.md",
    
    3: "[image]-image-urls.md",
    4: "[image]-image+prompt.md",
    5: "[image]-prompt.md",
    
    6: "[video]-prompt.md",
    7: "[video]-image-url.md",
    8: "[video]-image-urls.md",
    9: "[video]-video-url.md",
    10: "[music]-minimax.md"
}

INPUT_MD_FILE = TEMPLATES[TEMPLATE_INDEX]
MODEL_ID = "grok-imagine-video-1.5-preview"
DEVELOPER = "x-ai"
SCHEMA_URL = "https://raw.githubusercontent.com/aimlapi/api-docs/refs/heads/main/docs/api-references/video-models/xAI/grok-imagine-video-1.5-preview.json"

# =================


def schema_generation(schema_slug):
    response = requests.post(
        "https://api.gitbook.com/v1/orgs/isNk5xPnJebTWJa4JCg5/openapi",
        headers={"Authorization":"Bearer gb_api_vVVR78hJXh8dkcFDRtgn9LBHtFTQUmWAkQDdXe8q","Content-Type":"application/json"},
        data=json.dumps({
            "slug": schema_slug,
            "source": {
                "url": SCHEMA_URL
            }
        }),
    )
    data = response.json()
    print(data)
    
    
def make_schema_public(schema_slug):
    response = requests.patch(
        f"https://api.gitbook.com/v1/orgs/isNk5xPnJebTWJa4JCg5/openapi/{schema_slug}",
        headers={"Authorization":"Bearer gb_api_vVVR78hJXh8dkcFDRtgn9LBHtFTQUmWAkQDdXe8q","Content-Type":"application/json"},
        data=json.dumps({
        "visibility": "public"
        })
    )
    data = response.json()
    print(data)  
      

def main():
    
    # привязываем уже существующую схему в Гитбуке
    schema_generation(MODEL_ID.lower().replace("/", "-").replace(".", "-"))
    
    # читаем исходный файл md-шаблона
    with open(INPUT_MD_FILE, "r", encoding="utf-8") as f:
        content = f.read()

    # последовательно выполняем замены в md-шаблоне
    replacements = {
        "MODEL_ID": MODEL_ID,
        "DEVELOPER": DEVELOPER,
        "SCHEMA_URL": SCHEMA_URL,
        "SCHEMA_SLUG": MODEL_ID.lower().replace("/", "-").replace(".", "-")
    }
    
    print(MODEL_ID.lower().replace("/", "-").replace(".", "-"))
    for key, value in replacements.items():
        content = content.replace(key, value)

    # имя выходного файла
    output_file = f"{MODEL_ID}.md"

    # сохраняем результат
    with open(output_file, "w", encoding="utf-8") as f:
        f.write(content)

    print(f"Готово! Создан файл: {output_file}")

if __name__ == "__main__":
    main()
```
